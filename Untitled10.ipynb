```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

# Load the dataset (Update file name after uploading)
file_name = "your_uploaded_file.xlsx"  # Change this to the correct filename
xls = pd.ExcelFile(file_name)

# Display sheet names
print("Sheet Names:", xls.sheet_names)

# Load the first sheet (Modify if needed)
df = xls.parse(sheet_name=xls.sheet_names[0])

# Show first few rows
print("Dataset Preview:\n", df.head())

# **1. Data Cleaning**
df.dropna(inplace=True)  # Remove missing values

# Convert the first column to datetime (Modify column name if needed)
df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0], errors='coerce')

# Remove invalid dates
df.dropna(subset=[df.columns[0]], inplace=True)

# Set the first column as index (assuming it's Date)
df.set_index(df.columns[0], inplace=True)

# **2. Probability & Statistics**
print("\nStatistical Summary:\n", df.describe())

# **3. Data Visualization - CPI Trends**
plt.figure(figsize=(12, 6))
sns.lineplot(data=df)
plt.title("Consumer Price Index (CPI) Trend Over Time")
plt.xlabel("Year")
plt.ylabel("CPI Index")
plt.legend(df.columns)
plt.grid(True)
plt.show()

# Boxplot for CPI Variability
plt.figure(figsize=(10, 5))
sns.boxplot(data=df)
plt.title("CPI Distribution Over Time")
plt.xlabel("CPI Categories")
plt.ylabel("Values")
plt.grid(True)
plt.show()

# **4. Store in SQLite Database for SQL Queries**
conn = sqlite3.connect("financial_data.db")
df.to_sql("CPI_Data", conn, if_exists="replace", index=True)

# **5. Run SQL Queries**
query = "SELECT * FROM CPI_Data LIMIT 5;"  # Fetch first 5 rows
df_sql = pd.read_sql(query, conn)
print("\nSQL Query Output:\n", df_sql)

# Close connection
conn.close()
```
